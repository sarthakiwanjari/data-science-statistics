```python
#immporting the basics library
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=x*2
```

```python
y
```

LINE CHART

```python
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()

```

Bar Chart

```python
plt.bar(x,y)

plt.title("Line chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()

```

```python
plt.bar(x,y, color='black')
```

scatter plot

```python
a=(1,5,8,6,3,7,9,4)
b=(10,55,21,64,85,33,44,28)
plt.scatter(a,b)
plt.title("Scatter Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
